```python
import matplotlib.pyplot as plt
import numpy as np 
import json
```

```python
f = open('./reporting_website/api/report.json')
data = json.load(f)
f.close()
```

```python
attacks = data["attacks"]
phys_attacks = []
attack_hit = []

for attack in attacks:
    if attack[0] == 'physical':
        phys_attacks.append(attack[2])
        if attack[1]:
            attack_hit.append(1)
        else:
            attack_hit.append(0)
```

```python
a = np.array(phys_attacks)
hist, bins = np.histogram(a, bins=range(0,19))
plt.xticks(range(0,19))
plt.xlim(0,19)
plt.hist(bins[:-1], bins=range(0,19), weights=hist)
```

```python
a = np.array(attack_hit)
hist, bins = np.histogram(a, bins=range(0,19))
plt.xticks(range(0,19))
plt.xlim(0,19)
plt.hist(bins[:-1], bins=range(0,19), weights=hist)
```

```python
plt.plot(phys_attacks)
plt.show()
```
